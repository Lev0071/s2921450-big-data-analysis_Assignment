```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

x = [5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y = [7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,5,1]

colors = [7,5,9,7,5,7,2,5,3,7,1,2,8,1,9,2,5,6,7,5]
sizes = [423,220,267,296,440,538,394,411,248,523,255,317,393,326,375,434,325,435,417,305]


#plt.scatter(x,y,s=100,c="green",edgecolor='black',linewidth=1,alpha=0.75) # marker = "X"
#plt.scatter(x,y,s=100,c=colors,edgecolor='black',linewidth=1,alpha=0.75) # marker = "X"
# plt.scatter(x,y,s=100,c=colors,edgecolor='black',cmap='Greens',linewidth=1,alpha=0.75) # marker = "X"

plt.scatter(x,y,s=sizes,c=colors,edgecolor='black',cmap='Greens',linewidth=1,alpha=0.75) # marker = "X"

cbar = plt.colorbar()
cbar.set_label('Satisfaction')

plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('preprocessed.csv')
low = data['LowestSalary']
high = data['HighestSalary']

#data['AverageSalary'] = data.iloc[:, [10,11].mean(axis=1)
data["AverageSalary"] = data.loc[:, ["LowestSalary","HighestSalary"]].mean(axis = 1)
size = data["AverageSalary"]

plt.scatter(high,low,edgecolor='black',c=size,cmap='summer',linewidth=1,alpha=0.75) # marker = "X"

cbar = plt.colorbar()
cbar.set_label('Average of salary range')

plt.xscale('log')
plt.yscale('log')

plt.title('Seek Job Adverts')
plt.xlabel('Highest Salary')
plt.ylabel('Lowest Salary')

# https://www.youtube.com/watch?v=zZZ_RCwp49g
# @ 19:02
```
